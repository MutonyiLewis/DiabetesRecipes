##Import libraries

```python
import pandas as pd
import numpy as np
```

##Import data

```python
gl_datapath = '/content/drive/MyDrive/Attachment Folder/RecipesWithGILimitsWithNormalization.csv'
gl_df = pd.read_csv(gl_datapath)
gl_df.head(4)
```

##Acccess relevant columns

```python
#Populate the variables
df = pd.DataFrame(gl_df)

##Initialize vars
HighGI=np.array([df['High GI(%)']])
LowGI = np.array([df['Low GI(%)']])


Average = np.array([(HighGI+LowGI)/2])

print(list(Average[0][0]))
```

##Add new Average Column to Dataframe


```python
df['Average GI(%)'] = list(Average[0][0])
df.head(5)
```

##Export the dataset

```python
# export the DataFrame to a CSV file
df.to_csv('RecipesWithAverageGI.csv', index=False)
```
