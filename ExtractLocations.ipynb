##Import Libraries

```python
import pandas as pd
import re
```

##Access dataset

```python
folderpath = 'drive/MyDrive/Attachment Folder/'
path = 'RecipesWithAverageGI.csv'

df = pd.read_csv(folderpath+path, encoding='cp1252')
df.head(5)
```

##Use RE library to extract location Keywords

```python
# Create a new column 'location'
df['location'] = ''

# Iterate over the rows in the dataset
for index, row in df.iterrows():

    # Get the keywords from the column 'about'
    keywords = row['about'].split()

    # Check if the keywords are present in the list of keywords
    if 'across' in keywords :
        df.loc[index, 'location'] = 'Kenya'
    elif 'swahili' in keywords:
        df.loc[index, 'location'] = 'Coast'
    elif 'coastal' in keywords:
        df.loc[index, 'location'] = 'Coast'
    elif 'coast' in keywords:
        df.loc[index, 'location'] = 'Coast'
    elif 'taita' in keywords:
        df.loc[index, 'location'] = 'Coast'
    elif 'urban' in keywords:
        df.loc[index, 'location'] = 'Nairobi'
    elif 'nubian' in keywords:
        df.loc[index, 'location'] = 'Nairobi'
    elif 'western' in keywords:
        df.loc[index, 'location'] = 'Western'
    elif 'luhya' in keywords:
        df.loc[index, 'location'] = 'Western'
    elif 'luo' in keywords:
        df.loc[index, 'location'] = 'Nyanza'
    elif 'kikuyu' in keywords:
        df.loc[index, 'location'] = 'Central'
    elif 'embu' in keywords:
        df.loc[index, 'location'] = 'Embu'
    elif 'borana' in keywords:
        df.loc[index, 'location'] = 'Borana'
    elif 'Kamba' in keywords:
        df.loc[index, 'location'] = 'Eastern'
    elif 'kalenjin' in keywords:
        df.loc[index, 'location'] = 'Rift'
    elif '' in keywords:
        df.loc[index, 'location'] = 'Kenya'


# Save the updated dataset
df.to_csv('dataset_with_location.csv', index=False)

df.head(20)
```
