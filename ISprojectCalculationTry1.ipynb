```python
path = 'drive/MyDrive/Attachment Folder/'
trainpath = 'RecipesImp.csv'
```

```python
##Import libraries
import csv
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(path+trainpath, encoding='cp1252')
df.head(5)
```

```python
#Fill F Glycaemic Index column
#Calculate the available carbohydrates (AC) in the food
df['AC'] = df['carbohydrates(g)'] - df['fibre(g)']

#Calculate Estimated Sugar in recipes
df['Estimated Sugar'] = df['AC'] - (df['F_factor_est'] * df['fibre(g)'])

#Calculate the carbohydrate content that will significantly affect blood sugar (CHO)
df['CHO'] = df['AC'] - (0.5 * df['Estimated Sugar']) - (2 * df['proteins(g)']) - (0.5 * df['fat(g)'])

#Calculate the GI column
df['GI'] = 100 * (df['CHO'] / 100)
```

```python
df = pd.read_csv(path+trainpath, encoding='cp1252')
```

```python
#Fill F Glycaemic Index column
#Calculate the available carbohydrates (AC) in the food
df['AC'] = df['carbohydrates(g)'] - df['fibre(g)']

#Calculate Estimated Available Carbohydrates (EACA) in recipes
df['EACA'] = df['AC'] - (0.95 * df['fibre(g)'])

#Calculate the carbohydrate content that will significantly affect blood sugar (CHO)
df['CHO'] = df['AC'] - (0.5 * df['EACA']) - (2 * df['proteins(g)']) - (0.5 * df['fat(g)'])

#Calculate the GI column
df['GI'] = 100 * (df['CHO'] / 100)
```
