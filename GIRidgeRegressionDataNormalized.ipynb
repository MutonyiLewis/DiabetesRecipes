


##Import Libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

##Preview the data


```python
gl_datapath = '/content/drive/MyDrive/Attachment Folder/FoodsWithGIs.csv'
gl_df = pd.read_csv(gl_datapath)
gl_df.head(10)
```

##Scale the data

```python
df_dataframe = pd.DataFrame(gl_df)
df_min_max_scaled = df_dataframe.copy()

##Apply normalization to all columns
column = 'Calories(kcal)'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

column = 'Total Carbohydrate (g)'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

column = 'Dietary Fiber (g)'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

column = 'Protein (g)'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

column = 'Fat (g)'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

# view normalized data
display(df_min_max_scaled)
```

##Train test split

```python
# Load the training data
X = df_min_max_scaled.drop(columns=['Id','Food ID','Sugars','Glucose','Mixed meal','Gl (%)', 'GL','High Gl (%)', 'Low GI (%)','High GL'])
y = df_min_max_scaled['High Gl (%)']

##Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

##Define hyperparameters

```python
# Define the hyperparameter grid
param_grid = {
    "alpha": [0.01, 0.1, 1, 10, 100]
}
```

##Initialize Model

```python
model = Ridge()
```

##Perform Grid search

```python
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

##Get the best model

```python
best_model = grid_search.best_estimator_
```

##Model evaluation

```python
y_pred = best_model.predict(X)
score = np.mean(np.square(y - y_pred))
print("Mean squared error on test set:", score)
```
